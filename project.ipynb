#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
#Function to normalize a column (z-score normalization)
def normalize_column(column):
    return (column - column.mean()) / (column.std())

Fuenlabrada = ds_cleant[ds_cleant['level5'] == 'Fuenlabrada']
Alcorcon = ds_cleant[ds_cleant['level5'] == 'Alcorcón']
Getafé = ds_cleant[ds_cleant['level5'] == 'Getafe']
Leganés = ds_cleant[ds_cleant['level5'] == 'Leganés']

Fuenlabrada_normalized = normalize_column(Fuenlabrada['price'])
Alcorcon_normalized = normalize_column(Alcorcon['price'])
Getafé_normalized = normalize_column(Getafé['price'])
Leganés_normalized = normalize_column(Leganés['price'])

plt.figure(figsize=(12, 8))
plt.hist(Fuenlabrada_normalized, bins=30, alpha=0.5, label='Fuenlabrada', color='blue')
plt.hist(Alcorcon_normalized, bins=30, alpha=0.5, label='Alcorcón', color='orange')
plt.hist(Getafé_normalized, bins=30, alpha=0.5, label='Getafe', color='green')
plt.hist(Leganés_normalized, bins=30, alpha=0.5, label='Leganés', color='red')

plt.legend()
plt.title('Normalized Price Distribution by Population')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


```

The normalized price histograms reveal a classic positively skewed distribution across all neighborhoods, where most properties cluster below the mean while a smaller number of high-priced outliers create a long right tail, indicating that real estate markets typically have many moderately-priced options but prices rise dramatically in the premium segment.

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in Exercise 14) (★★★★)

Execute the following cell, and then start coding in the next one. You must implement a simple code that transforms the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
from ipyleaflet import Map, basemaps

# Map centered on south west part of Madrid
map = Map(center = (40.35, -3.8), zoom = 11.5, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
import numpy as np
from ipyleaflet import Map, basemaps, CircleMarker, WidgetControl
from ipywidgets import HTML

colors = {
    'Fuenlabrada': 'blue',
    'Leganés': 'red',
    'Getafe': 'green',
    'Alcorcón': 'black'
}

for idx, row in subset_south_belt.iterrows():
    try:
        lat = float(str(row['latitude']).replace(',', '.'))
        lon = float(str(row['longitude']).replace(',', '.'))
        if not np.isnan(lat) and not np.isnan(lon):
            popup_text = f"Address: {row['address']}<br>Price: {row['price']}<br>Surface: {row['surface']} m²"
            popup_widget = HTML(popup_text)
            circle = CircleMarker(
                location=(lat, lon),
                radius=3,
                color=colors.get(row['level5'], 'gray'),
                fill_color=colors.get(row['level5'], 'gray'),
                fill_opacity=0.6,
                popup=popup_widget
            )
            map.add_layer(circle)
    except Exception as e:
        print(f"Error at index {idx}: {e}")
        continue

legend_html = """
<div style="background-color: white; padding: 10px; border-radius: 5px;">
<h4>Population Legend</h4>
<ul>
    <li><span style="color: blue;">Fuenlabrada</span></li>
    <li><span style="color: red;">Leganés</span></li>
    <li><span style="color: green;">Getafe</span></li>
    <li><span style="color: black;">Alcorcón</span></li>
</ul>
</div>
"""

legend = HTML(legend_html)
legend_control = WidgetControl(widget=legend, position='bottomleft')
map.add_control(legend_control)
map
```
